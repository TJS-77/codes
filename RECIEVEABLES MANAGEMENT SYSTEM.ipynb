```python
#IMPORTING NECESSARY PACKAGES
import sqlite3                    
from tkinter import *            


# CREATING DATABASE 'ACCOUNT_REC' AS A CLASS
class DB:                         
    def __init__(self):           
        self.conn = sqlite3.connect("Account_rec.db")  
        self.cur = self.conn.cursor()   
        self.cur.execute(             
            "CREATE TABLE IF NOT EXISTS AR (cid INTEGER PRIMARY KEY, cname TEXT, credit INTEGER, duration INTEGER)") 
        self.conn.commit()  

    def __del__(self):          
        self.conn.close()  

    def view(self):       
        self.cur.execute("SELECT * FROM AR") 
        rows = self.cur.fetchall()  
        return rows

    def insert(self, cname, credit, duration):  
        self.cur.execute("INSERT INTO AR VALUES (NULL,?,?,?)", (cname, credit, duration,)) 
        self.conn.commit()
        self.view()

    def update(self, cid, credit, duration):   
        self.cur.execute("UPDATE AR SET credit=?, duration=? WHERE cid=?", (credit, duration, cid,))
        self.conn.commit()
        self.view()

    def delete(self, cid):                 
        self.cur.execute("DELETE FROM AR WHERE cid=?", (cid,))
        self.conn.commit()
        self.view()


# FUNCTIONALITIES FOR GUI BUTTONS AND DISPLAY AREA
def row_select(event):
    global selected_tuple
    cid = display.curselection()[0] 
    selected_tuple = display.get(cid) 
    cname.delete(0, END)               
    cname.insert(END, selected_tuple[1])
    credit.delete(0, END)
    credit.insert(END, selected_tuple[2])
    duration.delete(0, END)
    duration.insert(END, selected_tuple[3]) 

def view_cmnd():         
    display.delete(0, END)   
    for row in db.view():  
        display.insert(END, row) 

def add_cmnd():          
    db.insert(cname_text.get(), credit_integer.get(), duration_integer.get())
    display.insert(END, (cname_text.get(), credit_integer.get(), duration_integer.get())) 
    view_cmnd()

def delete_cmnd(): 
    db.delete(selected_tuple[0]) 
    view_cmnd()

def update_cmnd():
    db.update(selected_tuple[0], credit_integer.get(), duration_integer.get()) 
    view_cmnd()

    
# CREATING OBJECT 'db' OF CLASS 'DB'
db = DB()


# GUI
window = Tk()
window.title("ACCOUNTS RECIEVABLE SYSTEM")


# DIALOUGE BOX LABELS
l1 = Label(window, text="DEBTER NAME") 
l1.grid(row=0, column=0) 

l2 = Label(window, text="CREDIT AMOUNT")
l2.grid(row=1, column=0)

l3 = Label(window, text="CREDIT DURATION MONTHS")
l3.grid(row=2, column=0)


# DIALGUE BOX
cname_text = StringVar()
cname = Entry(window, textvariable=cname_text) 
cname.grid(row=0, column=1)

credit_integer = IntVar() 
credit = Entry(window, textvariable=credit_integer)
credit.grid(row=1, column=1)

duration_integer = IntVar() 
duration = Entry(window, textvariable=duration_integer)
duration.grid(row=2, column=1)


# DISPLAY AREA
display = Listbox(window, height=25, width=50) 
display.grid(row=4, column=0, rowspan=5, columnspan=2) 
display.bind('<<ListboxSelect>>', row_select)


# BUTTONS
b1 = Button(window, text="View all Records", width=13, command=view_cmnd, fg = "black", bg = "white") 
b1.grid(row=4, column=3)

b2 = Button(window, text="Add Record", width=13, command=add_cmnd, fg = "white", bg = "green")
b2.grid(row=5, column=3)

b3 = Button(window, text="Update Record", width=13, command=update_cmnd, fg = "white", bg = "orange")
b3.grid(row=6, column=3)

b4 = Button(window, text="Delete Record", width=13, command=delete_cmnd, fg = "white", bg = "red")
b4.grid(row=7, column=3)

b5 = Button(window, text="Close", width=13, command=window.destroy, fg = "white", bg = "black")
b5.grid(row=8, column=3)

window.mainloop() # END
```
